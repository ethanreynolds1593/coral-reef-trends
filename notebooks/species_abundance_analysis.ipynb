```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
species_count_data = pd.read_csv("../data/processed_data/CREMP_SCOR_Summaries_2023_Counts.csv")
```

```python
# 1. Species Abundance by Date
# Select a subset of species (modify as needed)
selected_species = ['Acropora_cervicornis', 'Acropora_palmata', 'Cladocor_arbuscula', 'Montastraea_cavernosa']

# Prepare data for species abundance over time
species_by_date = species_count_data[['Date'] + selected_species]

# Set Date as index for better visualization
species_by_date.set_index('Date', inplace=True)

# Plot the species abundance by date
plt.figure(figsize=(12, 7))
species_by_date.plot(kind='line', figsize=(12, 6), marker='o', colormap='tab10')
plt.title('Species Abundance by Date', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Species Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Species')
plt.tight_layout()
plt.savefig("../plots/species_abundance_by_date.png")
plt.show()
plt.close()
```
