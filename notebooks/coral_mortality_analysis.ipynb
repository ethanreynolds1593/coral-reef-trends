```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("../data/processed_data/CREMP_SCOR_RawData_2023.csv")
```

```python
# 1. Coral Mortality by Site (Bar Plot)
mortality_columns = ['Percent_old_mortality', 'Percent_recent_mortality']

# Prepare data for mortality by site
mortality_by_site = data[mortality_columns].copy()
mortality_by_site['Site'] = data['Site_name']

# Calculate the average mortality by site
mortality_by_site = mortality_by_site.groupby('Site').mean().reset_index()

# Plot the mortality by site
plt.figure(figsize=(12, 7))
mortality_by_site.set_index('Site')[mortality_columns].plot(kind='bar', stacked=True, color=['salmon', 'lightblue'], figsize=(12, 7))
plt.title('Coral Mortality by Site', fontsize=16)
plt.xlabel('Site', fontsize=12)
plt.ylabel('Mortality Percentage (%)', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Mortality Type')
plt.savefig("../plots/coral_mortality_by_site.png")
plt.show()
plt.close()
```

```python
# 2. Coral Size Distribution (Histogram)
plt.figure(figsize=(12, 7))
sns.histplot(data['Diameter_cm'], bins=30, kde=True, color='teal')
plt.title('Coral Size Distribution (Diameter)', fontsize=16)
plt.xlabel('Diameter (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.savefig("../plots/coral_size_distribution.png")
plt.show()
plt.close()
```

```python
# 3. Coral LTA by Species (Scatter Plot)
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Diameter_cm', y='LTA_cm2', data=data, hue='sciName', palette='Set2', marker='o')
plt.title('Coral LTA vs Diameter by Species', fontsize=16)
plt.xlabel('Diameter (cm)', fontsize=12)
plt.ylabel('LTA (cm²)', fontsize=12)
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.savefig("../plots/coral_lta_by_species.png")
plt.show()
plt.close()
```
