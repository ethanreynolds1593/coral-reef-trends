```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create output folder
os.makedirs("../plots", exist_ok=True)

# Load dataset
df = pd.read_csv("../data/processed_data/CREMP_Stations_2023.csv")
```

```python
# --- Plot 1: Station Map ---
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df.drop_duplicates("StationID"), x="lonDD", y="latDD",
                hue="Subregion", style="Habitat", s=100)
plt.title("Station Locations by Subregion and Habitat")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("../plots/station_locations_map.png")
plt.show()
plt.close()
```

```python
# --- Plot 2: Histogram of Depths ---
plt.figure(figsize=(12, 7))
sns.histplot(df["Depth_ft"], bins=15, kde=True, color="royalblue")
plt.title("Distribution of Survey Depths")
plt.xlabel("Depth (feet)")
plt.ylabel("Number of Stations")
plt.tight_layout()
plt.savefig("../plots/depth_distribution.png")
plt.show()
plt.close()
```

```python
# --- Plot 3: Histogram of Transect Lengths ---
plt.figure(figsize=(12, 7))
sns.histplot(df["length_m"], bins=10, kde=True, color="seagreen")
plt.title("Distribution of Station Transect Lengths")
plt.xlabel("Transect Length (meters)")
plt.ylabel("Number of Stations")
plt.tight_layout()
plt.savefig("../plots/transect_length_distribution.png")
plt.show()
plt.close()
```

```python
# --- Plot 4: Number of Stations per Year ---
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x="First_Year_surveyed", palette="viridis")
plt.title("Number of Stations First Surveyed by Year")
plt.xlabel("First Year Surveyed")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("../plots/stations_per_year.png")
plt.show()
plt.close()
```

```python
# --- Plot 5: Heatmap of Habitat x Subregion ---
heatmap_data = df.groupby(["Habitat", "Subregion"]).size().unstack(fill_value=0)
plt.figure(figsize=(12, 7))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Station Density: Habitat  Subregion")
plt.xlabel("Subregion")
plt.ylabel("Habitat")
plt.tight_layout()
plt.savefig("../plots/station_density_heatmap.png")
plt.show()
plt.close()
```
