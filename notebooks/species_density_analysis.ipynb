```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
density_data = pd.read_csv("../data/processed_data/CREMP_SCOR_Summaries_2023_Density.csv")
```

```python
# 1. Species Density by Date
# Select a subset of species (modify as needed)
selected_species = ['Acropora_cervicornis', 'Acropora_palmata', 'Cladocora_arbuscula', 'Montastraea_cavernosa']

# Prepare data for species density by site
density_by_site = density_data[['Site_name'] + selected_species]

# Set Site as index for better visualization
density_by_site.set_index('Site_name', inplace=True)
# Prepare data for species density over time
density_by_date = density_data[['Date'] + selected_species]

# Set Date as index for better visualization
density_by_date.set_index('Date', inplace=True)

# Plot the species density by date
plt.figure(figsize=(12, 7))
density_by_date.plot(kind='line', figsize=(12, 6), marker='o', colormap='tab10')
plt.title('Species Density by Date', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Species Density', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Species')
plt.tight_layout()
plt.savefig("../plots/species_density_by_date.png")
plt.show()
plt.close()
```
