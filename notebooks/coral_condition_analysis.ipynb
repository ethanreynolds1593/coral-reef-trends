```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
condition_data = pd.read_csv("../data/processed_data/CREMP_SCOR_Summaries_2023_ConditionCounts.csv")
```

```python
# 1. Coral Condition by Site (Stacked Bar Plot)
condition_columns = ['BLH', 'PBL', 'PAL', 'BBD', 'WBD', 'WPX', 'WPL', 'DSD', 'SCT', 'RTL', 'RBD', 'YBD', 'UNK']

# Prepare data for conditions by site
condition_by_site = condition_data[condition_columns].copy()
condition_by_site['Site'] = condition_data['Site_name']

# Calculate the total count for each condition by site
condition_by_site = condition_by_site.groupby('Site').sum()

# Plot the coral conditions by site
plt.figure(figsize=(12, 7))
condition_by_site.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='Set3')
plt.title('Coral Condition by Site', fontsize=16)
plt.xlabel('Site', fontsize=12)
plt.ylabel('Condition Count', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Condition Type')
plt.tight_layout()
plt.savefig("../plots/coral_condition_by_site.png")
plt.show()
plt.close()
```

```python
# 2. Coral Condition by Species (Stacked Bar Plot)
condition_by_species = condition_data[condition_columns].copy()
condition_by_species['Species'] = condition_data['sciName']

# Calculate the total count for each condition by species
condition_by_species = condition_by_species.groupby('Species').sum()

# Plot the coral conditions by species
plt.figure(figsize=(12, 7))
condition_by_species.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='Set2')
plt.title('Coral Condition by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Condition Count', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Condition Type')
plt.tight_layout()
plt.savefig("../plots/coral_condition_by_species.png")
plt.show()
plt.close()
```
