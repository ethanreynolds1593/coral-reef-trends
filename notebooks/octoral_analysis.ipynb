```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("../data/processed_data/CREMP_OCTO_RawData_2023.csv")
```

```python
# 1. Distribution of octocoral heights (Histogram)
plt.figure(figsize=(12, 7))
sns.histplot(data['Height_cm'], kde=True, color='skyblue', bins=20)
plt.title('Distribution of Octocoral Heights', fontsize=16)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.savefig("../plots/octocoral_height_distribution.png")
plt.show()
plt.close()
```

```python
# 2. Count of Octocoral Species per Site (Bar Plot)
species_count = data.groupby('sciName')['Height_cm'].count().reset_index(name='Count')
species_count = species_count.sort_values(by='Count', ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x='Count', y='sciName', data=species_count, palette='viridis')
plt.title('Count of Octocoral Species per Site', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Species', fontsize=12)
plt.grid(True)
plt.savefig("../plots/octocoral_species_count.png")
plt.show()
plt.close()
```

```python
# 3. Trends of Octocoral Heights over Time (Line Plot)
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year

# Aggregate average height by year
yearly_height = data.groupby('Year')['Height_cm'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(x='Year', y='Height_cm', data=yearly_height, marker='o', color='coral')
plt.title('Octocoral Height Trends Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Height (cm)', fontsize=12)
plt.grid(True)
plt.savefig("../plots/octocoral_trends_over_time.png")
plt.show()
plt.close()
```
